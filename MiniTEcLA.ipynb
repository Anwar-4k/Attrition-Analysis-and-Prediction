```python
def remove_outliers(HR_Attrition, threshold=3):
    mean = np.mean(HR_Attrition)
    std = np.std(HR_Attrition)
    z_scores = np.abs(HR_Attrition - mean) / std
    filtered_data = HR_Attrition[z_scores <= threshold]
    return filtered_data
cleaned_data = remove_outliers(HR_Attrition)
print(cleaned_data)
cleaned_data.head()
cleaned_data.describe()
```
